# Example of Data Distribution

### Uniform Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
values = np.random.uniform(-10,10,100000)
plt.hist(values,30)
plt.show()
```

## Normal / Gaussian 

Visualize the probablity density function 

```python
from scipy.stats import norm
```

```python
x = np.arange(-3,3,0.001)
plt.plot(x, norm.pdf(x))
```

Generate some random numbers with a normal distribution. "mu" is the desired mean, "sigma" is the standard deviation

```python
import numpy as np 
import matplotlib.pyplot as plt

mu = 5.0
sigma = 2.0

value  = np.random.normal(mu, sigma, 10000)
plt.hist(value, 50)
plt.show()
```

# Exponential PDF / "Power Law"

```python
from scipy.stats import expon
import matplotlib.pyplot as plt

x = np.arange(0,10,0.001)
plt.plot(x, expon.pdf(x))
```

# Binomial prob mass func

Pmf deals with discrete data

```python
from scipy.stats import binom
import matplotlib.pyplot as plt

n , p = 10, 0.5

x = np.arange(0,10,0.001)
plt.plot(x, binom.pmf(x,n,p))
```

# Poisson Probablity mass Function

Example: My website gets on average 500 visits per day. What's the odds of getting 550

```python
from scipy.stats import poisson
import matplotlib.pyplot as plt

mu = 500

x = np.arange(400, 600, 0.5)
plt.plot(x,poisson.pmf(x,mu))
```
